```python
import numpy as np
import import_ipynb
from scipy.fftpack import fft, fftfreq, ifft
import matplotlib.pyplot as plt
from Transformaciones_tiempo_frecuencia import Espectro, señal
from random import gauss
from Correlación_parcial import lcorr
```

```python
N = 10000
n = np.arange(N)
F = 60; Fs = 1200
x = 2* np.cos(2*np.pi*(F/Fs)*n+np.pi/4) + 15
ruido = np.random.normal(0,10,N)
x1 = x + ruido
```

```python
X1 = fft(x1)/N
p1 = np.arange(len(X1))
```

```python
plt.plot(p1*Fs/N,abs(X1))
plt.ylim(0,1.2)
plt.show()
```
