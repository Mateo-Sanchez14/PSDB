    Punto I) Para los siguientes pares de funciones: 
    1. Muestrear las funciones, adoptando valores adecuados para t y 𝑓𝑠.
    2. Determinar la convolución discreta entre las funciones muestreadas.
    3. Determinar la correlación discreta entre las funciones muestreadas.

    NOTA:
    Los comandos convolve() y correlate() de Numpy no tienen en cuenta el índice de comienzo de la convolución y correlación respetivamente. Considere esta situación para crear vectores de tiempo correspondiente.

### Apartado 1

```python
Fs = 50
s = np.arange(-10,10,1/Fs)
T, K, a = 5, 1, 1
h1 = np.piecewise(s, [abs(s) <= T, (s < -T) & (s > T)],  [lambda s: K, lambda s: 0])
g1 = np.piecewise(s, [s >= 0, s < 0],  [lambda s: np.exp(-a*s), lambda s: 0])
```

```python
plt.plot(s,h1)
plt.plot(s,g1, '--')
plt.xlabel('t / seg')
plt.ylabel('h1(t) - g1(t)')
plt.legend(('h1(t)','g1(t)'),prop={'size':15},ncol=3,bbox_to_anchor=(0.8,1.2))
plt.grid (True)
plt.show()
```

```python
#Convolución:
y1 = np.convolve(h1,g1)
t0 = s[0] + s[0]
Ny = len(h1) + len(g1) - 1
tconv = np.arange(t0,t0 + Ny/Fs,1/Fs)
yconv1 = y1/Fs

#Correlación:
r1 = np.correlate(g1,h1,'full')
t0 = s[0] - s[len(s)-1]
Ny = len(h1) + len(g1) - 1
tcorr = np.arange(t0,t0 + Ny/Fs,1/Fs)
corr1 = y1/Fs



plt.plot(tconv, yconv1)
plt.plot(tcorr, corr1 , '--')
plt.xlabel('t / seg')
plt.ylabel('h2(t) - g2(t)')
plt.legend(('h2(t)','g2(t)'),prop={'size':15},ncol=3,bbox_to_anchor=(0.8,1.2))
plt.grid (True)
plt.show()
```

### Apartado 2

```python
Fs2 = 50
t2 = np.linspace(-10,10-(1/Fs2), 20 * Fs2)

a = 1
b = 5

h2 = np.piecewise(t2, [t2 > 0, t2 < 0], [lambda t2: np.exp(-a*t2), lambda t2: 0])
g2 = np.piecewise(t2, [t2 > 0, t2 < 0], [lambda t2: np.exp(-b*t2), lambda t2: 0])
```

```python
plt.plot(t2,h2)
plt.plot(t2,g2, '--')
plt.xlabel('t / seg')
plt.ylabel('h2(t) - g2(t)')
plt.legend(('h2(t)','g2(t)'),prop={'size':15},ncol=3,bbox_to_anchor=(0.8,1.2))
plt.xlim (-1,10)
plt.grid (True)
plt.show()
```

```python
CV2 = np.convolve(h2, g2)

ny0 = t2[0] + t2[0] #La convolución empieza en la suma de los comienzos
Ny = len(t2)/Fs2 + len(t2)/Fs2 - 1 #Suma de longitudes - 1: Número de elementos del vector
ny = np.linspace(ny0, ny0+Ny, 40 * Fs2 - 1)


CR2 = np.correlate(h2, g2, "full")

nr0 = t2[0]- t2[len(t2)-1]
Nr = len(t2)/Fs2 + len(t2)/Fs2 -1
nr = np.linspace(nr0, nr0 + Nr, 40 * Fs2 - 1)
```

```python
plt.plot(ny, CV2)
plt.plot(nr, CR2 , '--')
plt.xlabel('t / seg')
plt.ylabel('h2(t) - g2(t)')
plt.legend(('h2(t)','g2(t)'),prop={'size':15},ncol=3,bbox_to_anchor=(0.8,1.2))
plt.grid (True)
plt.show()
```

    Punto III) La respuesta al impulso de un sistema, muestreada a 100 Hz, es:

    h[ n ] = {0 .41 .67 .27 0 .18 0.3 .12 0 .08 .14 .06 0 .04 .06 .03 0 .02.03 .01}. Encontrar y dibujar la respuesta del sistema a una entrada xd[ n ], formada por los muestreos a 100 Hz de x(t)=1.5·sen(2·pi·7·t)+0.8·sen(2·pi·14·t)+0.6·sen(2·pi 21·t).

```python
plt.plot(n3,x3_n)
plt.xlim(0,100)
plt.xlabel('n')
plt.ylabel('x3_n')
plt.grid (True)
plt.show()
```

```python
plt.stem(h3)
plt.xlabel('n')
plt.ylabel('h3')
plt.show()
```

```python
CV3 = np.convolve(x3_n, h3)
ncv0 = n3[0] + nh[0] 
Ncv3 = len(n3) + len(nh) - 1
ncv3 = np.arange(ncv0, ncv0 + Ncv3)
```
