```python
import numpy as np
import matplotlib.pyplot as plt
```

Esta función reconstruye una señal a partir de sus muestras. Los argumentos necesarios son: 
n: longitud del vector tiempo.
Fs: frecuencia de muestreo temporal (Hz). Dado que no es posible obtener una señal contínua de t, se representará la señal reconstruida emplesando una frecuencia fs muy alta, de modo que no se aprecie la separación entre punto y punto. 
Usualmente Fs >> Fs1.
datos: este es un vector que contiene las muestras de la señal a reconstruir.
Fs1: Frecuenia de muestreo empleada para obtener las muestras (Hz).

```python
def h(n,Fs,datos,Fs1):
    y = np.ndarray(n); n1 = len(datos)
    for i in np.arange(n): # i es el contador de tiempos
        z = 0
        for j in np.arange(n1):
            if j == (i/Fs)*Fs1:  # (i/Fs) es el tiempo
                z = z + datos[j]
            else:
                z = z + datos[j]*np.sin(2*np.pi*(Fs1/2)*(i/Fs-j/Fs1))/(2*np.pi*(Fs1/2)*(i/Fs-j/Fs1))
        y[i] = z
    return y
```
