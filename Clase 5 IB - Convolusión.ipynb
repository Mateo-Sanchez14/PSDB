### Convolución


Si la convolución es continua, pueden usarse funciones.

Pero si la convolución es discreta, se usarán vectores: x(n) y h(n).

Se usan para encontrar la respuesta de un filtro.

### Ejemplo 1:

x -> [ *1* 1 1 ]  
0,1,2

h -> [ *5* 4 3 2 1 ]

```python
nx = np.arange(-7,10)
x = lambda nx: np.piecewise(nx,[(nx>=0)&(nx<=2)],[lambda nx: 1, lambda nx: 0])
nh = nx
h = lambda nh: np.piecewise(nh,[(nh>=0)&(nh<5)],[lambda nh: 5-nh, lambda nh: 0])
```

```python
plt.plot(figsize=(10,4))
plt.plot(nx,x(nx),'o',ms=6)
plt.plot(nh,h(nh),'ko',ms=10,mfc='none')
plt.xticks(nx)
plt.legend(('x(n)','h(n)'),prop={'size':15},ncol=2,bbox_to_anchor=(0.6,1.2))
plt.grid(True)
plt.show()
```

```python
#PLEGADO (FOLDING)

plt.figure(figsize=(20,4))
plt.subplot(121)
plt.plot(nx,x(nx),'o',ms=6)
plt.plot(nh,h(nh),'ko',ms=10,mfc='none')
plt.vlines(0,0,5,'r',linestyles='dashed')
plt.xticks(nx)
plt.legend(('x(n)','h(n)'),prop={'size':15},ncol=2,bbox_to_anchor=(0.8,1.3))
plt.grid(True)


plt.subplot(122)
plt.plot(nx,x(nx),'o',ms=6) #no sería -nx??
plt.plot(nh,h(-nh),'ko',ms=10,mfc='none')
plt.vlines(0,0,5,'r',linestyles='dashed')
plt.xticks(nx)
plt.legend(('x(n)','h(-n)'),prop={'size':15},ncol=2,bbox_to_anchor=(0.8,1.3))
plt.grid(True)
plt.show()
```

### h(k-n) significa que desplazamos h(-n) k lugares. Si k es negativo los desplazamientos se produce hacia la izquierda y viceversa. Si k = -1 nos movemos hacia la izquierda un lugar y en esta situación no hay puntos en común entre x(nx) y h(-1-nh).

```python
plt.figure(figsize=(20,4))
plt.subplot(121)
plt.plot(nx,x(nx),'o',ms=6)
plt.plot(nh,h(-1-nh),'ko',ms=10,mfc='none')
plt.vlines(0,0,5,'r',linestyles='dashed')
plt.text(1.5,4,s='k=-1 → Conv(-1)=0',fontsize=13)
#Cualquier valor de los puntos * los azules que valen 0 = 0
plt.xticks(nx)
plt.legend(('x(n)','h(-1-n)'),prop={'size':15},ncol=2,bbox_to_anchor=(0.8,1.3))
plt.grid(True)


plt.subplot(122)
plt.plot(nx,x(nx),'o',ms=6)
plt.plot(nh,h(0-nh),'ko',ms=10,mfc='none')
plt.text(1.5,4,s='k=0 → Conv(0)=1*5=5',fontsize=13)
plt.vlines(0,0,5,'r',linestyles='dashed')
plt.xticks(nx)
plt.legend(('x(n)','h(0-n)'),prop={'size':15},ncol=2,bbox_to_anchor=(0.8,1.3))
plt.grid(True)
plt.show()
```

### Implementación de la función "np.convolve" de numpy

### Digamos que j es un índice o contador que recorre los elementos de los vectores nx, x, nh y h. j inicia en 0 y termina en 16

```python
j = np.arange(len(nx))
nx, x(nx), nh, h(nh)
```

```python
j = 0
nx[j], x(nx)[j], nh[j], h(nh)[j]
```

```python
j = 5
nx[j], x(nx)[j], nh[j], h(nh)[j]
```

### Llamo "y" a la convolución entre x y h

```python
y = np.convolve(x(nx), h(nh)) # Calculamos la convolución
ny0 = nx[0] + nh[0] #La convolución empieza en la suma de los comienzos
Ny = len(nx) + len(nh) - 1 #Suma de longitudes - 1: Número de elementos del vector
ny = np.arange(ny0, ny0+Ny)
```

```python
plt.figure(figsize=(8,4))
plt.plot(ny,y,'o',color='purple',ms=6)
plt.xlim(-2,10)
plt.xticks(np.arange(-3,10))
plt.title('Convol(x,h)',fontsize=13)
plt.grid(True)
plt.show()
```

### Tarea:
### Ejemplo2: Encuentra la convolución entre x y h, donde nx = (2,3,4), x = (1,1,1), nh = (-3,-2,-1,0,1), h = (5,4,3,2,1)


```python
NX = np.arange(2,5)
X = np.array ([1,1,1])
NH = np.arange (-3,2)
H = np.array ([5,4,3,2,1])
```

```python
plt.plot(figsize=(10,4))
plt.plot(NX,X,'o',color="red",ms=6)
plt.plot(NH,H,'ko',ms=10,mfc='none')
plt.xticks(NX)
plt.legend(('X(N)','H(N)'),prop={'size':15},ncol=2,bbox_to_anchor=(0.6,1.2))
plt.grid(True)
plt.show()
```

```python
Y = np.convolve(X, H) # Calculamos la convolución
NY0 = NX[0] + NH[0] #La convolución empieza en la suma de los comienzos
NY = len(NX) + len(NH) - 1 #Suma de longitudes - 1: Número de elementos del vector
n_y = np.arange(NY0, NY0+NY)
```

```python
plt.figure(figsize=(8,4))
plt.plot(n_y,Y,'o',color='orange',ms=6)
plt.xlim(-2,10)
plt.xticks(np.arange(-3,10))
plt.yticks(np.arange(0,13,1))
plt.title('Convol(x,h)',fontsize=13)
plt.grid(True)
plt.show()
```

### Ejemplo3:


```python
n = np.arange (-2,6)
delta = lambda n: np.piecewise (n, [n == 0] , [lambda n: 1, lambda n: 0])

x_n = (1/2 * delta(n)) + (2*delta(n-1))
h_n = delta (n) + delta (n-1) + delta (n-2)
```

```python
plt.figure(figsize=(20,4))
plt.subplot(121)
plt.plot(n,x_n,'o',ms=6)
plt.xticks(n)
plt.legend(('x(n)'),prop={'size':15},ncol=2,bbox_to_anchor=(0.8,1.3))
plt.grid(True)


plt.subplot(122)
plt.plot (n,h_n,'ko',ms=10)
plt.xticks(n)
plt.legend(('h(n)'),prop={'size':15},ncol=2,bbox_to_anchor=(0.8,1.3))
plt.grid(True)
plt.show()
```

```python
y3 = np.convolve(x_n, h_n) # Calculamos la convolución
ny03 = n[0] + n[0] #La convolución empieza en la suma de los comienzos
Ny3 = len(n) + len(n) - 1 #Suma de longitudes - 1: Número de elementos del vector
ny3 = np.arange(ny03, ny03+Ny3)
```

```python
plt.figure(figsize=(8,4))
plt.plot(ny3,y3,'o',color='green',ms=6)
plt.xlim(-2,10)
plt.xticks(np.arange(-3,10))
plt.title('Convol(x,h)',fontsize=13)
plt.grid(True)
plt.show()
```
