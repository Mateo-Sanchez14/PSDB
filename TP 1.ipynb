**PUNTO 2**

Aparatado 1: SEÑAL DE PRESIÓN

```python
C1 = 80
Fc1 = 90/60 #pulsos por minuto
w1 =  np.array ([1,2,3,4])*2*np.pi*Fc1
A1 = [20.000,10.0240,3.5560,0.9800]
O1 = [0,-0.5655,1.0053,-2.1363] 
t1 = np.linspace (0,5,1001)
```

```python
cos0 = A1[0] * np.cos ((w1[0]*t1)-O1[0])
cos1 = A1[1] * np.cos ((w1[1]*t1)-O1[1])
cos2 = A1[2] * np.cos ((w1[2]*t1)-O1[2])
cos3 = A1[3] * np.cos ((w1[3]*t1)-O1[3])

P = C1 + cos0 + cos1 + cos2 + cos3

#GRÁFICA
plt.figure(figsize=(8,6))
plt.plot(t1,P)
plt.grid (True)
plt.xlabel('Tiempo',fontsize=12)
plt.ylabel('Presión',fontsize=12)
plt.show()
```

```python
#Para hacerlo iterando

H1 = np.zeros (len (t1))
for i in range (len(t1)):
    H1[i] = C1 + sum (A1 * np.cos( w1*t1[i] - O1))
    
#GRÁFICA
plt.figure(figsize=(8,6))
plt.plot(t1,H1)
plt.grid (True)
plt.xlabel('Tiempo',fontsize=12)
plt.ylabel('Presión',fontsize=12)
plt.show()
```

```python
H2 = np.zeros (len (t2))
for i in range (len(t2)):
    H2[i] = C2 + sum ( A2 * np.cos(w2*t2[i]) + B2 * np.sin (w2*t2[i]) )
    
#GRÁFICA
plt.figure(figsize=(8,6))
plt.plot(t2,H2, color="r")
plt.grid (True)
plt.xlabel('Tiempo',fontsize=12)
plt.title ("ECG", fontsize=20)
plt.show()
```
