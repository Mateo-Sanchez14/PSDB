    EJERCICIO I

    Una se√±al anal√≥gica contiene frecuencias hasta 10 kHz.

    a) ¬øQu√© rango de frecuencias de muestreo permite una reconstrucci√≥n exacta de esta se√±al a partir de sus muestras?

    b) Sup√≥n que se muestrea la se√±al a una frecuencia de 8kHz. Analiza qu√© sucede con la frecuencia de 5kHz.
    
    c) Repite el an√°lisis del punto anterior para el caso de la frecuencia de 9kHz.

### Apartado a) 

Digamos que su m√°xima frecuencia es Fmax = 10kHz.

Sabemos que la m√°xima frecuencia a reconstruir sin ambiguedad es Fs/2.

Para evitar Aliasing => Fs/2 > Fmax => Fs > 2 * Fmax

**Es decir que como m√≠nimo se requiere que la frecuencias de muestreo sean mayores a 20 [kHz]**

### Apartado b) 

Si se muestrea una se√±al de 5 kHz con una frecuencia de muestreo de 8 kHz. Tenemos que:

La curva tendr√° menos de dos puntos en la definici√≥n de su se√±al. Por lo que en lugar de verse la se√±al que deber√≠a ser podr√≠amos encontrarnos con frecuencia alias (se produce ALIASING).

Esto se consigue restando la frecuencia de muestreo por la frecuencia de la se√±al:

#### Fa = Fs - F0 = 3 [kHz]

Por lo que en lugar de verse la se√±al de 5 [kHz] se ver√≠a la de 3 [kHz]

La frecuencia relativa de la misma se obtiene:

5000/8000 = 5/8 = 1 - 3/8 ---> fa = 3/8

Fa1 = fa * Fs = 3 KHz

### Apartado c) 

Si se muestrea una se√±al de 9 kHz con una frecuencia de muestreo de 8 kHz. Tenemos que:

La curva tendr√° menos de dos puntos en la definici√≥n de su se√±al. Por lo que en lugar de verse la se√±al que deber√≠a ser podr√≠amos encontrarnos con frecuencia alias.

Esto se consigue restando la frecuencia de muestreo por la frecuencia de la se√±al:

#### Fa = F1 - Fs = 1 [kHz]

Por lo que en lugar de verse la se√±al de 9 [kHz] se ver√≠a la de 1 [kHz]

La frecuencia relativa de la misma se obtiene:

**f relativa = F1 / Fs**

9000/8000 = 5/8 = 1 + 1/8 ---> frec de alias2 = fa2 = 1/8

Fa2 = fa2 * Fs = 8 KHz

    b) Suponga que se muestrea la se√±al con una frecuencia de muestreo fs=5000Hz, ¬øcu√°l ser√° la se√±al discreta en el tiempo x(n) obtenida luego del muestreo?

    c) ¬øQu√© se√±al anal√≥gica ya (t) se podr√° reconstruir a partir del muestreo?

    d) Grafique xa (t), x(n) y ya (t).

ùë•_a (t) = 3cos(2000œÄùë°) + 5sin (6000œÄùë°) + 10 cos(12000œÄùë°)

ùë•_a (t) = 3cos(1000 * 2œÄùë°) + 5sin (3000 * 2œÄùë°) + 10 cos(6000 * 2œÄùë°)

Fs = 5000 Hz .   x(n) ?

Reemplazo "t" por "n/Fs"

x(n) = 3 cos((1000/5000)* 2œÄn) + 5 sin((3000/5000)* 2œÄn) + 10 cos((6000/5000)* 2œÄn)

-- Se hace eso porque deben ser menores a 1/2 --

x(n) = 3*cos[(1/5)*2*œÄ*n] + 5*sin[2*œÄ*n-(2/5)*2*œÄ*n] + 10 * cos[2*œÄ*n+(1/5)*2*œÄn]

x(n) = 13*cos[(1/5)*2*œÄ*n] - 5*sin[(2/5)*2*œÄ*n]        ‚úî

Con esa frecuencia de muestreo, ser√≠a esta se√±al la que estar√≠amos muestreando.

### Apartado c) 

y_a(t)? La se√±al an√°logica reconstuida a partir del muestreo.

Reemplazo en x(n) ---> "n" por "t * Fs"

y_a(t) = 13*cos[(1/5)*5000*2*œÄ*t] - 5*sin[(2/5)*5000*2*œÄ*t]

y_a(t) = 13*cos[(1000*2*œÄ*t] - 5*sin[(2000*2*œÄ*t]        ‚úî

### Apartado d) 

Graficar: x_a(t) , x(n) e y_a(t)

### Representaci√≥n de la se√±al x_a (t)
##### ùë•_a (t) = 3cos(1000 * 2œÄùë°) + 5sin (3000 * 2œÄùë°) + 10 cos(6000 * 2œÄùë°)

```python
#REPRESENTO x_a(t)
t = np.arange(0,1,0.00001)
x_a = 3*np.cos(1000*2*np.pi*t) + 5*np.sin(3000*2*np.pi*t) + 10 * np.cos(6000*2*np.pi*t)
```

```python
plt.figure(figsize=(15,4))
plt.plot(t,x_a)
plt.grid()
plt.xlabel('t [seg]', fontsize=14)
plt.legend(('Se√±al x_a(t)',0),prop={'size':14},ncol=2,bbox_to_anchor=(0.2,1.01))
plt.xlim(0,0.003)
plt.show()
```

### Representaci√≥n de la se√±al x(n)
##### x(n) = 13*cos[(1/5)*2*œÄ*n] - 5*sin[(2/5)*2*œÄ*n]

```python
#REPRESENTO x(n)
Fs = 5000
n = np.arange(len(t))
x = 13*np.cos((1/5)*2*np.pi*n) - 5*np.sin((2/5)*2*np.pi*n)
```

```python
plt.figure(figsize=(15,4))
plt.plot(n/Fs,x,'o', c='purple')
plt.plot(t,x_a)
plt.grid()
plt.xlabel('t [seg]', fontsize=14)
plt.legend(('Se√±al x(n)','Se√±al x_a(t)'),prop={'size':14},ncol=2,bbox_to_anchor=(0.2,1.01))
plt.xlim(0,0.003)
plt.show()
```

### Representaci√≥n de la se√±al y_a(t)
##### y_a(t) = 13*cos[(1000*2*œÄ*t] - 5*sin[(2000*2*œÄ*t]

```python
#REPRESENTO y_a(t)
y_a = 13*np.cos(1000*2*np.pi*t) - 5*np.sin(2000*2*np.pi*t)
```

```python
plt.figure(figsize=(15,4))
plt.plot(t,x_a,c='r')
plt.plot(n/Fs,x,'o', c='purple')
plt.plot(t,y_a,c='k')
plt.grid()
plt.xlabel('t [seg]', fontsize=14)
plt.legend(('Se√±al x_a(t)','Se√±al x(n)','Se√±al y_a(t)'),prop={'size':14},ncol=2,bbox_to_anchor=(0.2,1.01))
plt.xlim(0,0.003)
plt.show()
```

### Representaci√≥n de la se√±al y_a(t) con la funci√≥n de reconstrucci√≥n:
##### x(n) = 13*cos[(1/5)*2*œÄ*n] - 5*sin[(2/5)*2*œÄ*n]

```python
# En este caso Fs debe ser MUY grande porque se est√° trabajando con una frecuencia de muestreo Fs1 muy alta (5000Hz)
xrec1 = h(1000,50000,x[0:1000],5000) # h(n, Fs, datos, Fs1) ... Fs >> Fs1
t1 = np.arange(len(xrec1))/50000
```

```python
plt.figure(figsize=(15,4))
plt.plot(t1,xrec1,c='y')
plt.plot(n/Fs,x,'o', c='purple')
plt.plot(t,x_a,c='r')
plt.grid()
plt.xlabel('t [seg]', fontsize=14)
plt.legend(('Se√±al xrec1(t)','Se√±al x(n)','Se√±al x_a(t)'),prop={'size':14},ncol=2,bbox_to_anchor=(0.2,1.01))
plt.xlim(0,0.003)
plt.show()
```

### Apartado b)
