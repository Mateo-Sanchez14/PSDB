```python
import numpy as np
```

Esta función devuelve la envolvente de la señal "a" empleando una ventana móvil de ancho "window_size". En cada ventana se calcula el valor rms definido como: 

```python
def envolvente_rms(a, window_size):
    a2 = np.power(a,2)
    window = np.ones(window_size)/float(window_size)
    return np.sqrt(np.convolve(a2, window, 'valid'))
```
