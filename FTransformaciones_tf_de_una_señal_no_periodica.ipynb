### Función para obtener el espectro de una señal de t-discreto y aperiódica

Esta función permite obtener el espectro de una señal de t-discreto no periódica. Para ello es necesario ingresar 3 elementos: los vectores (xinput) y (tinput) que representan la señal cuyo espectro se desea obtener y su correspondiente vector de tiempos en (seg) y la frecuencia de muestreo Fs que se empleó para obtener xinput. Los vectores xinput y tinput deben tener dimensión 1xN.

```python
def EspecCont(tinput,xinput,Fs):
    n = tinput*Fs; Fsw = 1000; Nw = 2000; k = np.arange(Nw); 
    w = np.pi*(k-Nw/2)/Fsw; espec = np.ndarray(Nw,dtype=complex)
    for i in k:
        espec[i] = np.sum(xinput * np.exp(-1j*w[i]*n))
    R = np.vstack((w,espec)).T
    return R
```
