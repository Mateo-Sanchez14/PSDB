<h2 style="color:red;" align="center" > USO DEL IMPULSO UNITARIO PARA REPRESENTAR SEÑALES DE T. DISCRETO = CONVOLUCIÓN </h2>

### El sistema discreto= Transformación trabaja con muestras:



### x(n) = función de entrada
### y(n) = función de salida 

### x(n), y(n) = valores de las funciones en el instante ACTUAL de tiempo

### x(n-1), y(n-1)= valores en el instante anterior
### x(n+1), y(n+1)= valores en el instante posterior  (lleva a modos de trabajo offline) 

<h2 style="color:lightgreen;"> SISTEMAS ESTÁTICOS O SIN MEMORIA </h2>

#### Si su salida para cada valor de la vble. independiente SÓLO depende de la entrada ACTUAL, es decir de x(n). Ejemplo: un resistor

#### y1(n)=a.x(n) + 1 ---> Sí es estático, depende de x(n)
#### y2(n)=b.x(n) + n. x^2(n)  ---> No es estático puesto que n está como producto 



### NO es estático cuando aparece el factor n, x(n-1), x(n+1)

<h2 style="color:green;"> SISTEMAS DINÁMICOS O CON MEMORIA </h2>

#### Si su salida para cada valor de n depende de las entradas anteriores, es decir x(n-1) ó x(n-2). Por ejemplo el acumulador es un sistema con memoria.

#### y3(n)= sum (x(n-k))  Tiene memoria infinita

<h2 style="color:red;"> SISTEMAS INVARIANTES Y VARIANTES EN EL TIEMPO </h2>

#### Es invariante en el tiempo si su Transformación no cambia con el tiempo. Es decir, si la entrada se retrasa k unidades y realizo la transformación, la salida también se retrasa k unidades. 
#### --> Su transformación característica no cambia con el tiempo 




<h2 style="color:purple;"> SISTEMAS LINEALES Y NO LINEALES </h2>

#### Un sistema es lineal sii se cumplen las propiedades de linealidad de la transformación y se llega a que: 




<h2 style="color:orange;"> SISTEMAS CAUSALES Y NO CAUSALES </h2>

#### Sistema causal: la salida actual depende de las entradas (actual y anteriores)

#### Sistema no causal: la salida actual depende de AL MENOS una entrada posterior / futura. Estas señales se trabajan con procesamiento offline, ya que la computadora no puede trabajar con datos que no conoce.




<h2 style="color:blue;"> SISTEMAS ESTABLES E INESTABLES </h2>

#### El sistema es estable si para cualquier entrada finita, el sistema produce una salida finita. 
#### Teniendo en cuenta el concepto de energía de una serie:



#### Es decir es ESTABLE cuando la Energía luego de la transformación, sigue siendo finita. 


<h1 style="color:RED;"> Al trabajar con sistemas lineales, se los puede caracterizar en el dominio del tiempo y en el dominio de la frecuencia.
SLID: SISTEMAS LINEALES INVARIANTES EN EL TIEMPO</h1>

## (1) Caracterización en el dominio del tiempo : CONVOLUCIÓN

## Ecuación de diferencias con coeficientes constantes



### ak y br son coeficientes reales, ak acompaña a las salidas y br a las entradas

### N es el mayor retraso temporal de y

<h1 style="color:RED;"> SI N=0, SE OBTIENE UNA RESPUESTA AL IMPULSO FINITA (FIR) </h1>

### Por lo tanto el caso más simple es cuando N=0, la salida solo depende del momento actual.




### -> La salida no depende de salidas anteriores, sí de x(n) y x(n-k)

<h1 style="color:RED;"> SI N=!0, SE OBTIENE UNA RESPUESTA AL IMPULSO INFINITA (IIR) </h1>

## N indicará la cantidad de condiciones iniciales que se necesitan

```python
#EJEMPLO1

## Impulso Unitario
n1=np.arange(0,10)
Delta1=np.piecewise(n1, [n1==0], [lambda n1:1, lambda n1:0])


M=5
y1=(1/M)*sum(Delta1[n1-i] for i in range(0,M))
```

```python
plt.figure(figsize=(8,4))
plt.stem(n1,y1)
plt.xlim(-2,10)
plt.hlines(0.2,-0.01,4.1,linestyle="dashed")
plt.title('RESPUESTA AL IMPULSO DE UN FILTRO DE PROMEDIOS MÓVILES ORDEN 5',fontsize=13)
plt.grid(True)
plt.show()
```

### Los valores de los coeficientes (Cr) del vector respuesta al impulso unitario tienen un valor de 1/5, todos iguales entre ellos. Ésto no siempre sucede, pueden ser diferentes

```python
#EJEMPLO 2

n2=np.arange(0,21)
Unistep2=np.piecewise (n2,[n2>=0], [lambda n2:1, lambda n2:0])
nh=np.arange(5)
h=np.array([1,1,1,1,1])/5  #vector de los coeficientes


plt.plot(n2,Unistep2,"o",color="black")
plt.plot(nh,h,"o",color="blue",mfc="none",ms=10)
plt.xlim(-1,21)
plt.ylim(0,1.1)
plt.xticks(np.arange(0,21,5))
plt.xlabel("$n(muestras)$", fontsize=15)
plt.legend(("Unistep","h"), prop={"size":12},
          loc="center right", bbox_to_anchor=(0.9,0.5))
plt.grid(True)
plt.show()
```

```python
y=np.convolve(Unistep2,h)
n0=nh[0]+n2[0]
Ny=len(nh)+len(n2)-1
ny=np.arange(n0,n0+Ny)
```

```python
plt.plot(ny,y,"o",color="r")
plt.xlabel("$n$")
plt.ylabel("$y=convol(Unistep,h)$")
plt.ylim(0,1.1)
plt.grid(True)
plt.show()
```

```python
#2) 

nb=np.arange(-10,10)
Unistepb= np.piecewise (nb,[nb>=0], [lambda nb:1, lambda nb:0])

nh=np.arange(len(nb))
h=np.ones(len(nb)) #Vector de unos, son los coeficientes cr=br/a0
```

```python
plt.figure(figsize=(15,4))
plt.plot(nh,h,"o",label=("coeficientes"))  #Gráfica del vector de los coeficientes
plt.plot(nb,Unistepb,"ko",label=("impulso"),mfc="None")  #mdf=none= PUNTITOS VACÍOS
plt.legend(fontsize=15)
plt.grid(True)
plt.show()
```

###  Encuentro la respuesta del sistema: es la convolución

```python
y2=np.convolve(h,Unistepb)
n02=nh[0]+nb[0]  #donde empieza la convolución
Ny2=len(nh) + len(nb) -1     #cuántos puntos tendrá 
ny2=np.arange(n02,n02+Ny2)
```

```python
plt.figure(figsize=(15,4))
plt.plot(ny2,y2,"ro",label="y(n)")
plt.xlabel("n/muestras")
plt.ylabel("y=conv(h,Unistep)")
plt.legend(fontsize=15)
plt.grid(True)
plt.show()
```
