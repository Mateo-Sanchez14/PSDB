### PUNTO 1 - A

𝑥1(𝑛) = Cos (2𝜋𝑛/12 + 𝜋/6) if 0≤𝑛<50; 

𝑥1(𝑛) = 1/2 Sin (𝜋𝑛/7 −𝜋/4) if −50≤𝑛<0 

```python
n1 = np.arange(-70,100,1,dtype=int) #como es x(n) solo debemos comtemplar valores enteros de paso
f1 = np.piecewise (n1, [(0>n1) & (n1>=-50) , (n1>=0) & (n1<50)] , [lambda n1: (np.sin( ((np.pi*n1)/7) - (np.pi/4) )) / 2 , lambda n1: np.cos( ((2*np.pi*n1)/12) + (np.pi/6) ) ])


#GRÁFICA
plt.plot(n1,f1,color='green')
plt.grid (True)
plt.xlim(-55,55)
plt.xlabel('n',fontsize=12)
plt.title('Gráfica 1 "A"',fontsize=8)
plt.show()
```

La parte real de 𝑥2(𝑡)=∑ C𝑘 𝑒^(𝑗2𝜋𝑘𝐹0𝑡)

Con 𝑐0 = 5, 𝑐1 = 𝑐4 = 10 e^(iπ/3), 𝑐2 = 𝑐3 = 5 e^(iπ/4) y C−𝑘 = C𝑘̅ (complejo conjugado de  𝑐𝑘) y 𝐹0=1/5 𝐻𝑧. 

```python
C = np.ndarray(9,dtype=complex)
C[4] = 5
C[8] = 10*cmath.exp(0+((1j)*np.pi/3))
C[5] = C[8]
C[0] = np.conjugate(C[8])
C[3] = C[0]
C[6] = C[7] = 5*cmath.exp(0+((1j)*np.pi/4))
C[1] = C[2] = np.conjugate(C[6])
F0 = 1/5
print(C[0],C[5])

x8 = np.arange(-100,100,0.1)
F8 = np.arange (len (x8),dtype=float)
for i in range(len(x8)):
    F8[i]=np.real(sum((C[j+4]*np.float_power(np.exp(np.complex(0,((2*np.pi)*j*F0))),x8[i])) for j in range(-4,5)))

plt.figure(figsize=(24,8))
plt.xlim(-100,100)
plt.grid (True)
plt.plot(x8,F8)
plt.title('Señal 1b)')
plt.xlabel('t/seg')
plt.ylabel('x2(t)')
plt.show()
```
